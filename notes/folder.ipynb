```python

from pathlib import Path


root = Path('/home/lysz210/PycharmProjects/simple-local-gallery/upload/')
```

```python
files = [{"folder": p.parent, "filename": p.name} for p in root.rglob('*.jpg', case_sensitive=False)]
files
```

```python
import pandas as pd


data = pd.DataFrame(files)

data
```

```python
grouped = data.groupby('folder').agg(photos=('filename', list)).reset_index()

groupted_folders = grouped.to_dict(orient='records')
groupted_folders
```
